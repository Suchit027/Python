```python
import matplotlib.pyplot as plt
import numpy

t = numpy.arange(0, 10, 0.1)
y1 = numpy.sin(t)
y2 = numpy.cos(t)
plt.plot(t, y1, 'b--', t, y2, 'r-.', linewidth=4.0)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('trigonometric graphs', fontsize=15)
plt.legend(['Sin', 'Cosine'], loc=5)
plt.grid(True)
plt.show()
```
